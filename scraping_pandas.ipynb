```python
population = tables[17]
population_df = population.iloc[:,0:2]
population_df.columns = ['Year','Population']
population_df.dropna(subset=['Population'], inplace=True)
population_df.set_index('Year')
#population_df.dtypes
#population_df.drop(labels='2016[68]')
#population_df
population_df2 = population_df.truncate(1,115, copy = False)
population_df2
```

```python
population_df2.to_csv(r'population.csv', index = False)
```

```python
population_df3=pd.to_numeric(population_df2['Population']).astype(int)
#population_df3=pd.to_numeric(population_df2['Year']).astype(int)
population_df3
```

```python
import matplotlib
```

```python
xlabels = ('')
plot1 = population_df3.plot.bar(x='Year', y='Population')
plot1.set_title("Population Growth from 1901 to 2015", fontsize='xx-large')
plot1.set_ylabel("Population in Millions")
plot1.set_xticklabels(xlabels)
```

```python
ethnic = tables[2]
ethnic_df = ethnic.iloc[:,0:2]

ethnic_df.columns = ['Location','Residents']
ethnic_df.set_index('Location')
ethnic_df
#ethnic_df.dtypes
```

```python
ethnic_df2 = ethnic_df.truncate(2,18, copy = False)
ethnic_df3 = ethnic_df2.set_index('Location')
ethnic_df3
```

```python
ethnic_df2.to_csv(r'ethnic.csv', index = False)
```

```python
xlabels = ('')
plot2 = ethnic_df3.plot.pie(x='Location', y='Residents')
plot2.set_title("Ethnicity of Residents Migrating to Australia", fontsize='xx-large')
plot2.set_ylabel("Ethnic Groups")
plot2.set_xticklabels(xlabels)
```

```python
distribution = tables[4]
distribution_df = distribution.iloc[:,0:2]
distribution_df
#distribution_df.dtypes
distribution_df.columns = ['State','Population']
distribution_df2 = distribution_df.set_index('State')
distribution_df2
```

```python
distribution_df.to_csv(r'distribution.csv', index = False)
```

```python
xlabels = ('')
plot2 = distribution_df2.plot.pie(x='State', y='Population')
plot2.set_title("Breakdown of Population per State", fontsize='xx-large')
plot2.set_ylabel("State")
plot2.set_xticklabels(xlabels)
```
