```python
import requests
wiki = requests.get('https://en.wikipedia.org/wiki/Demography_of_Australia').text

from bs4 import BeautifulSoup
soup = BeautifulSoup(wiki,'lxml')
print(soup.prettify())
```

```python
immi_table = soup.find('table', {'class':'wikitable plainrowheaders'})
immi_table
```

```python
links = immi_table.findAll('a')
links
```

```python
countries = []
for link in links:
    countries.append(link.get('title'))

print(countries)
```

```python
import pandas as pd 
df = pd.DataFrame()
df['Country'] = countries

df
```
